```python
# how many missing grades are across all exams in the dataset

exam['grade'].isnull().sum().sum()
```

```python
#filling the null value

exam_clean = exam.fillna(value = 0)

exam_clean
```

```python
exam_pass = exam["exam"]> 70
```

```python
# how many grades are passing greater than 70

len(exam_pass)
```

```python
# grouping the dataset by exam

exam_clean.groupby(['exam']).sum()
```

```python
# mean grade for all exams

exam_clean.groupby(['exam'])['grade'].agg(['mean','sum']) 
```
